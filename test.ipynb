```python
def initiate_langchain_model():
    """
    This function returns langchain azure open ai instance
    """
    # deployment_name = "gpt-35-16k"
    deployment_name = "gpt4-1106"
    model_name = "gpt-35-16k"
    temperature=0
    max_tokens = 800
    llm = AzureChatOpenAI(temperature=temperature,deployment_name=deployment_name, max_tokens = max_tokens)
    return llm
```

```python
def langchain_azure_response(query):
    """
    This function returns response answer of the question based on the context from langchain azure open ai 
    """
    llm = initiate_langchain_model()
    system_message = """You are a helpful ai assitant"""
    messages = [SystemMessage(content=system_message)]
    messages.append(HumanMessage(content=query))
    response_langchain_azure = llm(messages)
    return response_langchain_azure.content
```

```python

def extract_text(filepath):
    ext = filepath.rsplit('.', 1)[1].lower()
    chunks = []
    text = ""

    try:
        if ext == 'pdf':  
            with fitz.open(filepath) as doc:  
                # text = [] 
                for page in doc:  
                    chunks.append(page.get_text())
        # return text  
            # with fitz.open(filepath) as doc:  
            #     for page in doc:  
            #         text = page.get_text()  
            #         if text:  
            #             chunks.append(text)
        elif ext == 'txt':
            with open(filepath, 'r', encoding='utf-8') as file:
                text = file.read()
                chunks.append(text)
            
        elif ext == 'pptx':
            presentation = Presentation(filepath)
            for slide in presentation.slides:
                slide_text = ""
                for shape in slide.shapes:
                    if hasattr(shape, "text"):
                        slide_text += shape.text + '\n'
                if slide_text:
                    chunks.append(slide_text.strip())
                    
        elif ext == 'docx':
            doc = Document(filepath)
            for para in doc.paragraphs:
                text = para.text
                if text:
                    chunks.append(text)
                    
        else:
            text = textract.process(filepath).decode('utf-8')
            chunks.append(text)
            
    except FileNotFoundError:
        return f"File not found: {filepath}", ext
    except IOError:
        return f"Could not read file: {filepath}", ext
    except Exception as e:
        return f"An error occurred while extracting text: {e}", ext
    
    return chunks, ext
```
